- df.plot(kind='xxx')

### 1-01. Line plot

```python
# 사용할 DataFrame 준비
import pandas as pd
import numpy as np
np.random.seed(0)
df = pd.DataFrame(
    np.random.randn(100, 3),
    index = pd.date_range('2021-01-01', periods=100),
    columns=['You', 'I', 'We']).cumsum()
df.head()
```

```python
# 'line' 이 기본값이기 때문에 df.plot()으로 해도 동일
df.plot(kind='line')   
#df.plot.line() 
plt.grid()
plt.show()
```

```python
# bar 그래프
import seaborn as sns
iris = sns.load_dataset("iris")  # 붓꽂
# sepal : 꽃받침
# petal : 꽃잎

df = iris.groupby(iris.species).mean()
df.columns.name = 'feature'
df.plot(kind='bar', rot=0)
#df.plot.bar(rot=0)  # label 표시에 대한 것
plt.title('Mean')
```

### 2-01. 표준 정규분포(normal distribution) 

image.png

```python
import numpy as np

x = np.linspace(-5, 5, 101)   # x 정의
# 정규분포 : N(μ, σ/σ2)으로 표시
# 표준정규분포의 평균 μ는 0, 분산/표준편차 σ/σ2는 1 이다
y = (1 / np.sqrt(2 * np.pi)) * np.exp(- x ** 2 / 2 )
#display(x, y)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(8, 4))          # 플롯 사이즈 지정
plt.plot(x, y)                       
plt.xlabel("x")                      # x축 레이블 지정
plt.ylabel("y")                      # y축 레이블 지정
plt.grid()                           # 플롯에 격자 보이기
plt.title("Normal Distribution without scipy")     # 타이틀 표시
plt.legend(["N(0, 1)"])              # 범례 표시
plt.show()                           # 플롯 보이기
```

```python
import scipy.stats as stats

# stats.norm(평균, 표준편차).pdf(x) 를 사용하여 y 작성
y1 = stats.norm(0, 1).pdf(x)        
plt.figure(figsize=(8, 4))          # 플롯 사이즈 지정
plt.plot(x, y1, color="red")         # 선을 빨강색으로 지정하여 plot 작성          
plt.xlabel("x")                      # x축 레이블 지정
plt.ylabel("y")                      # y축 레이블 지정
plt.grid()                           # 플롯에 격자 보이기
plt.title("Normal Distribution with scipy.stats")     # 타이틀 표시
plt.legend(["N(0, 1)"])              # 범례 표시
plt.show()                           # 플롯 보이기
```

```python
import scipy.stats as stats

legend = []
y1 = stats.norm(0, 1).pdf(x)     
def norm (mu, std) :
    legend.append("N(" + str(mu) + ", " + str(std) + ")")
    return stats.norm(mu, std).pdf(x)        

plt.figure(figsize=(8, 4))          # 플롯 사이즈 지정

# 표준편차를 1 ~ 4로 지정함
for i in np.arange(1, 5):
    plt.plot(x, norm(0, i))          # plot 추가       
    
plt.xlabel("x")                      # x축 레이블 지정
plt.ylabel("y")                      # y축 레이블 지정
plt.grid()                           # 플롯에 격자 보이기
plt.title("Normal Distribution with scipy.stats")     # 타이틀 표시
plt.legend(legend)                   # 범례 표시
plt.show()                           # 플롯 보이기
```
